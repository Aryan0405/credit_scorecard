```python
bureau_agg = bureau_data.groupby('SK_ID_CURR').agg(
    BUREAU_LOAN_COUNT=('SK_ID_BUREAU', 'count'),
    BUREAU_AMT_CREDIT_SUM=('AMT_CREDIT_SUM', 'sum'),
    BUREAU_AMT_OVERDUE_SUM=('AMT_CREDIT_SUM_OVERDUE', 'sum'),
    BUREAU_AMT_OVERDUE_MAX=('AMT_CREDIT_SUM_OVERDUE', 'max'),
    BUREAU_DAYS_OVERDUE_MAX=('CREDIT_DAY_OVERDUE', 'max'),
    BUREAU_DELINQUENT_COUNT=('CREDIT_DAY_OVERDUE', lambda x: (x > 0).sum())
).reset_index()
```

```python
df = application_data.merge(bureau_agg, on='SK_ID_CURR', how='left')
print(df.shape)
print(df[['BUREAU_LOAN_COUNT','BUREAU_AMT_OVERDUE_SUM']].isnull().sum())
```

```python
df['BUREAU_NO_HISTORY'] = df['BUREAU_LOAN_COUNT'].isnull().astype(int)
```

```python
bureau_cols = ['BUREAU_LOAN_COUNT', 'BUREAU_AMT_CREDIT_SUM', 
               'BUREAU_AMT_OVERDUE_SUM', 'BUREAU_AMT_OVERDUE_MAX',
               'BUREAU_DAYS_OVERDUE_MAX', 'BUREAU_DELINQUENT_COUNT']

df[bureau_cols] = df[bureau_cols].fillna(0)
```

```python
print(df.isnull().sum()[df.isnull().sum() > 0].shape)
print(df['BUREAU_NO_HISTORY'].value_counts())
```

```python
split_threshold = df['DAYS_ID_PUBLISH'].quantile(0.80)

train = df[df['DAYS_ID_PUBLISH'] <= split_threshold].copy()
test = df[df['DAYS_ID_PUBLISH'] > split_threshold].copy()

print(train.shape, test.shape)
print(train['TARGET'].mean(), test['TARGET'].mean())
```

```python
df['DAYS_EMPLOYED'].describe()
```

```python
df['DAYS_EMPLOYED_ANOMALY'] = df['DAYS_EMPLOYED'].apply(lambda x: 1 if x == 365243 else 0)
df['DAYS_EMPLOYED'] = df['DAYS_EMPLOYED'].replace(365243, np.nan)
```

```python
df[df['DAYS_EMPLOYED'].isnull()].shape[0] / df.shape[0] * 100
```

```python
cols = ['DAYS_BIRTH', 'DAYS_EMPLOYED', 'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
train[cols].isnull().sum() / len(train) * 100
```

```python
train['EXT_SOURCE_1_MISSING'] = train['EXT_SOURCE_1'].isnull().astype(int)
train.groupby('EXT_SOURCE_1_MISSING')['TARGET'].mean()
```

```python
train.drop('EXT_SOURCE_1_MISSING', axis=1, inplace=True)
```

```python
from optbinning import OptimalBinning
```

```python
binning_dict = {}

for feature in cols:
    optb = OptimalBinning(name=feature, dtype="numerical", solver="cp")
    optb.fit(train[feature], train["TARGET"])

    train[feature + "_WoE"] = optb.transform(train[feature], metric="woe")
    test[feature + "_WoE"] = optb.transform(test[feature], metric="woe")

    binning_dict[feature] = optb
```

```python
woe_cols = [col + '_WoE' for col in cols]
train[woe_cols].head()
```

```python
train[woe_cols].isnull().sum()
```

```python
print(binning_dict['EXT_SOURCE_1'].binning_table.build())
```

```python
train[train['EXT_SOURCE_1_WoE'] == 0.0]['EXT_SOURCE_1'].isnull().sum()
```

```python
train[train['EXT_SOURCE_1_WoE'] == 0.0].shape[0]
```

```python
for feature in cols:
    bt = binning_dict[feature].binning_table.build()
    iv = bt['IV'].iloc[-1]
    print(f"{feature}: IV = {iv:.4f}")
```

```python
X_train = train[woe_cols]
y_train = train['TARGET']
X_test = test[woe_cols]
y_test = test['TARGET']
```

```python
from sklearn import linear_model
log_reg = linear_model.LogisticRegression(class_weight='balanced', max_iter=1000)
log_reg.fit(X_train, y_train)
```

```python
prediction_probabilities = log_reg.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score = roc_auc_score(y_test, prediction_probabilities)
print(f"ROC AUC Score: {roc_auc_score:.4f}")
```

```python
gini = 2 * roc_auc_score(y_test, prediction_probabilities) - 1
print(f"Gini Coefficient: {gini:.4f}")
```

```python
from scipy.stats import ks_2samp
```

```python
ks_statistic, p_value = ks_2samp(prediction_probabilities[y_test == 0], prediction_probabilities[y_test == 1])
```

```python
print(f"KS Statistic: {ks_statistic:.4f}, p-value: {p_value:.4f}")
```

Score = Offset + Factor × ln(Odds)

```python
PDO = 20
Factor = PDO / np.log(2)
offset= 600 - Factor * np.log(50)
print(f"Factor: {Factor:.4f}, Offset: {offset:.4f}")
```

```python
test['SCORE'] = offset + Factor * np.log((1 - prediction_probabilities) / prediction_probabilities)
test[['SK_ID_CURR', 'SCORE']].head()
```
